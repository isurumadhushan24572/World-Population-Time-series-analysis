### ADF Test [ Checking Stationary Part]

```python
from statsmodels.tsa.stattools import adfuller
```

Null Hypothesis :- Data set is Not Stationary

Alternative Hypothesis :- Data set is  Stationary

```python
result = adfuller(russian['Population'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print(f'Critical Values:{result[4]}')

if result[1] < 0.05 :
    print('Series is  stationary')
else:
    print('Series is Not stationary')
```

### KPSS Test

```python
from statsmodels.tsa.stattools import kpss
```

Null Hypothesis :- Data set is Stationary

Alternative Hypothesis :- Data set is not Stationary

```python
stats, p, lags, critical_value = kpss(russian['Population'])
print('p-value: %f' % p)
print('statistics: %s' % stats)
print('critical_value: %s' % critical_value)

if p < 0.05:
    print('Series is Not stationary')
else:
    print('Series is stationary')
```

```python
print(russian.shape)
train =russian.iloc[:-13]
test = russian.iloc[-13:]

print(train.shape,test.shape)
```

```python
from pmdarima import auto_arima
stepwise_fit = auto_arima(train['Population'],trace= True,suppress_warnings=True)
stepwise_fit.summary()
```

```python

from statsmodels.tsa.arima.model import ARIMA

# Assuming 'train' is your DataFrame containing the time series data
# and 'Population' is the column containing the population data

# Instantiate the ARIMA model with the order parameter (0, 2, 0)
model = ARIMA(train['Population'], order=(0, 2, 0))

# Fit the ARIMA model
model = model.fit()

# Print model summary
print(model.summary())
```

```python
# Plot residual errors
residuals = pd.DataFrame(model.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
# Build Model

model = ARIMA(train.Population, order=(0, 2, 0))  
fitted = model.fit()  
# Print model summary
print(fitted.summary())
# Forecast
fc = fitted.forecast(15)  # Forecast 15 steps ahead

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.xticks(rotation='vertical')  # Rotate x-axis labels vertically
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()

```

```python
start = len(train)
end = len(train) + len(test)-1
pred = fitted.predict(start=start,end=end,type='levels')
print(pred)
pred.index = russian.index[start:end+1]
```

```python
pred.plot(legend=True)
test['Population'].plot(legend=True)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt

# Calculate Mean Squared Error
MSE = mean_squared_error(pred, test['Population'])

# Calculate Root Mean Squared Error
RMSE = sqrt(MSE)

# Calculate Mean Absolute Percentage Error
MAPE = mean_absolute_error(test['Population'], pred) / test['Population'].mean() * 100

# Print the errors
print(f'Mean Squared Error: {MSE}')
print(f'Root Mean Squared Error: {RMSE}')
print(f'Mean Absolute Percentage Error: {MAPE}%')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit SARIMAX model
model = SARIMAX(train.Population, order=(0, 2, 0))
fitted_model = model.fit()

# Forecast
n_periods = 24
forecast = fitted_model.forecast(steps=n_periods)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train.Population, label='Training Data')
plt.plot(fitted_model.fittedvalues, color='red', label='Fitted Values')
plt.plot(pd.Series(forecast, index=test.index), color='green', label='Forecast')
plt.xticks(rotation='vertical')  # Rotate x-axis labels vertically
plt.title("ARIMA Model with SARIMAX")
plt.legend()
plt.show()
```
