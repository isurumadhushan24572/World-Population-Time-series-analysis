### ADF Test [ Checking Stationary Part]

```python
from statsmodels.tsa.stattools import adfuller
```

Null Hypothesis :- Data set is Not Stationary

Alternative Hypothesis :- Data set is  Stationary

```python
result = adfuller(china['Population'],regression='ct')
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print(f'Critical Values:{result[4]}')

if result[1] < 0.05 :
    print('Series is  stationary')
else:
    print('Series is Not stationary')
```

### KPSS Test

```python
from statsmodels.tsa.stattools import kpss
```

Null Hypothesis :- Data set is Stationary

Alternative Hypothesis :- Data set is not Stationary

```python
stats, p, lags, critical_value = kpss(china['Population'],'ct')
print('p-value: %f' % p)
print('statistics: %s' % stats)
print('critical_value: %s' % critical_value)

if p < 0.05:
    print('Series is Not stationary')
else:
    print('Series is stationary')
```

#### Finding No of defferences 

```python
import numpy as np, pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Import data
df = china

# Original Series
fig, axes = plt.subplots(4, 2, sharex=True)
axes[0, 0].plot(df.Population); axes[0, 0].set_title('Original Series')
plot_acf(df.Population, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.Population.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.Population.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df.Population.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.Population.diff().diff().dropna(), ax=axes[2, 1])

# 3rd Differencing
axes[3, 0].plot(df.Population.diff().diff().diff()); axes[3, 0].set_title('3rd Order Differencing')
plot_acf(df.Population.diff().diff().diff().dropna(), ax=axes[3, 1])

plt.show()
```

# ARIMA

```python
print(china.shape)
train =china.iloc[:-13]
test = china.iloc[-13:]
print(train.shape,test.shape)
```

```python

from statsmodels.tsa.arima.model import ARIMA

# Assuming 'train' is your DataFrame containing the time series data
# and 'Population' is the column containing the population data

# Instantiate the ARIMA model with the order parameter (1, 1, 1)
model = ARIMA(train['Population'], order=(1, 1, 1))

# Fit the ARIMA model
model = model.fit()

# Print model summary
print(model.summary())
```

```python
start = len(train)
end = len(train) + len(test)-1
pred = model.predict(start=start,end=end,type='levels')
print(pred)
pred.index = china.index[start:end+1]
```

```python
pred.plot(legend=True)
test['Population'].plot(legend=True)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt

# Calculate Mean Squared Error
MSE = mean_squared_error(pred, test['Population'])

# Calculate Root Mean Squared Error
RMSE = sqrt(MSE)

# Calculate Mean Absolute Percentage Error
MAPE = mean_absolute_error(test['Population'], pred) / test['Population'].mean() * 100

# Print the errors
print(f'Mean Squared Error: {MSE}')
print(f'Root Mean Squared Error: {RMSE}')
print(f'Mean Absolute Percentage Error: {MAPE}%')
```

#### Train the Model

```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'train' is your DataFrame containing the time series data
# and 'Population' is the column containing the population data

# Instantiate the ARIMA model with the order parameter (2, 2, 0)
model = ARIMA(train['Population'], order=(3, 1, 2))

# Fit the ARIMA model
model = model.fit()

# Print model summary
print(model.summary())
```

#### Predicting Model

```python
start = len(train)
end = len(train) + len(test)-1
pred = model.predict(start=start,end=end,type='levels')
print(pred)
pred.index = china.index[start:end+1]
```

```python
pred.plot(legend=True)
test['Population'].plot(legend=True)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt

# Calculate Mean Squared Error
MSE = mean_squared_error(pred, test['Population'])

# Calculate Root Mean Squared Error
RMSE = sqrt(MSE)

# Calculate Mean Absolute Percentage Error
MAPE = mean_absolute_error(test['Population'], pred) / test['Population'].mean() * 100

# Print the errors
print(f'Mean Squared Error: {MSE}')
print(f'Root Mean Squared Error: {RMSE}')
print(f'Mean Absolute Percentage Error: {MAPE}%')
```

```python
# Plot residual errors
residuals = pd.DataFrame(model.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```
